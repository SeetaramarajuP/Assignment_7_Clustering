# Crime Data

H Clustering

```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
crime=pd.read_csv('crime_data.csv')
crime
```

```python
def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return (x)
crime_norm=norm_func(crime.iloc[:,1:])
crime_norm.head()
```

```python
#   Another Method
from sklearn.preprocessing import MinMaxScaler
trans=MinMaxScaler()
data=pd.DataFrame(trans.fit_transform(crime.iloc[:,1:]))
data
```

```python
from scipy.cluster.hierarchy import linkage
import scipy.cluster.hierarchy as sch
z=linkage(crime_norm,method='average',metric='euclidean')
sch.dendrogram(z)
plt.title('Dendrogram')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
h_crime=AgglomerativeClustering(n_clusters=5,linkage='average',affinity='euclidean').fit(crime_norm)
cluster_labels=pd.Series(h_crime.labels_)
cluster_labels
```

```python
Crime_H=crime
Crime_H['Cluster']=cluster_labels
Crime_H
```

```python
Crime_H.iloc[:,1:].groupby(Crime_H.Cluster).mean()
```

K_ Mean 

```python
from sklearn.cluster import KMeans
fig=plt.figure(figsize=(10,8))
WCSS=[]
for i in range (1,11):
    clf=KMeans(n_clusters=i)
    clf.fit(crime_norm)                      # From H Cluster
    WCSS.append(clf.inertia_)
plt.plot(range(1,11),WCSS)
plt.title('Elbow Curve')
plt.plot()
```

```python
Kmodel=KMeans(n_clusters=5)
Kmodel.fit(crime_norm)
Kmodel.labels_
```

```python
clf.inertia_
```

```python
clf.cluster_centers_
```

```python
md=pd.Series(Kmodel.labels_)
Crime_K=crime
Crime_K['K_Cluster']=md
Crime_K=Crime_K.drop(['Cluster'],axis=1)
Crime_K
```

```python
Crime_K.iloc[:,1:].groupby(Crime_K.K_Cluster).mean()
```

DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
Crime_D=crime.iloc[:,1:5]
Crime_D
```

```python
Crime_D.values
```

```python
stscalar=StandardScaler().fit(Crime_D.values)
x=stscalar.transform(Crime_D.values)
x
```

```python
from sklearn.neighbors import NearestNeighbors
neigh=NearestNeighbors(n_neighbors=8)
nbrs=neigh.fit(x)
distances,indices=nbrs.kneighbors(x)
```

```python
distances=np.sort(distances,axis=0)
distances=distances[:,1]
plt.plot(distances)
```

```python
dbscan=DBSCAN(eps=1.30,min_samples=5)
dbscan.fit(x)
```

```python
dbscan.labels_
```

```python
c1=pd.DataFrame(dbscan.labels_,columns=['D_Cluster'])
c1
```

```python
Crime_D=pd.concat([Crime_D,c1],axis=1)
Crime_D
```
